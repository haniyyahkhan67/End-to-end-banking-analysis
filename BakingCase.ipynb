```python
# Value Counts for Categorical Features
category_cols = ['IAId','GenderId','BRId','Amount of Credit Cards','Nationality','Fee Structure','Loyalty Classification','Properties Owned']
for col in category_cols:
    print(f"Value counts for '{col}':")
    print(df[col].value_counts())
```

## 3. Data Visualization & Insights


```python
for i, col in enumerate(category_cols, start=1):
    plt.figure(figsize=(12, 6))  
    sns.countplot(data=df, x=col)
    plt.title(f"Count Plot for '{col}'")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
cols = ["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", 
         "Fee Structure", "Loyalty Classification", "Properties Owned"]

for i, col in enumerate(cols):
    plt.figure(i)
    sns.countplot(data=df, x=col, hue='GenderId')
    plt.title(f"Count Plot of {col} by GenderId")
    plt.xticks(rotation=45)
    plt.tight_layout()      
```

```python
numerical_col= ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
plt.figure(figsize=(12,32))
for i, col in enumerate(numerical_col):
    plt.subplot(len(numerical_col),1,i+1)
    sns.histplot(data=df,x=col,kde=True)
```

```python
corr_matrix=df[numerical_col].corr()
sns.heatmap(corr_matrix,annot=True,cmap='crest')
plt.title('Correlation Matrix Heatmap')
```
